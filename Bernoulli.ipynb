```python
import random

def bernoulli_rv_generator(num, p):
    numlist = []

    while num > 0:
        alpha = random.random()
        if 0 <= alpha <= (1-p):
            numlist.append(0)
        elif (1-p) <= alpha <= 1:
            numlist.append(1)
        num-=1
    
    return numlist
    
```

```python
bernoulli_rv_generator(1000, 0.7)
```
