```python
import random

def binomial_rv_generator(num, n, p):
    numlist = []
    
    while num > 0:
        num2 = n
        berlist = []
        
        while num2 > 0: 
            alpha = random.random()
            if 0 <= alpha <= (1-p):
                berlist.append(0)
            elif (1-p) <= alpha <= 1:
                berlist.append(1)
            num2 -= 1
   
        x = 0
    
        for y in berlist:
            x += y
    
        numlist.append(x)
        num -= 1
    
    return numlist
    
```

```python
binomial_rv_generator(1000, 10, 0.2)
```
