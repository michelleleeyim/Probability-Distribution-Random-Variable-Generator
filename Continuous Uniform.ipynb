```python
import random

def continuous_uniform_rv_generator(num, a, b):
    numlist = []

    while num > 0:
        alpha = random.random()
        x = (b - a) * alpha + a
        numlist.append(x)
        num -= 1
    
    return numlist
    
```

```python
continuous_uniform_rv_generator(1000, 1, 2)
```
