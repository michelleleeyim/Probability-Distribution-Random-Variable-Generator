```python
import random

def discrete_distribution_rv_generator(num):
    numlist = []

    while num > 0:
        alpha = random.random()
        if alpha < 0.2:
            numlist.append(1)
        elif 0.2 <= alpha < 0.5:
            numlist.append(2)
        elif 0.5 <= alpha < 0.9:
            numlist.append(3)
        elif 0.9 <= alpha <= 1:
            numlist.append(4)
        num-=1
    
    return numlist
    
```

```python
discrete_distribution_rv_generator(1000)
```
