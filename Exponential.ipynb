```python
import random
import math

def exponential_rv_generator(num, lam):
    numlist = []

    while num > 0:
        alpha = random.random()
        x = -1 * lam * math.log(1 - alpha)
        numlist.append(x)
        num-=1
    
    return numlist
    
```

```python
exponential_rv_generator(1000, 0.5)
```
