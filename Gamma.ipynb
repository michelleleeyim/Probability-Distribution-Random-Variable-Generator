```python
import random
import math

def gamma_rv_generator(num, a, b):
    numlist = []
    
    while num > 0:
        explist = []
        n = 0
        num2 = a
        
        
        while num2 > 0: 
            alpha = random.random()
            x = -1 * (1/b) * math.log(1 - alpha)
            explist.append(x)
            num2 = num2-1
    
        for y in explist:
            n = n + y
    
        numlist.append(n)
        num -= 1
    
    return numlist
    
```

```python
gamma_rv_generator(1000, 3, 2)
```
